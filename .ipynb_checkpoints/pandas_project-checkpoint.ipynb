```python
import pandas as pd
```

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_colwidth',None)
```

```python
def opener_of_files(link: str):
    with open(link, encoding='utf-8') as a:
        read_file = a.readlines()
        unpacker = [line.rstrip() for line in read_file]
        return unpacker           
```

```python
def results(arr):
    new_arr = []
    for i in arr:
        x = int(i)
        new_arr.append(x)
    return new_arr
    
```

```python
def table(frame,caption):
    return frame.style.set_caption(caption).set_table_styles([{
        'selector': 'caption',
        'props': [
            ('color', 'Black'),
            ('font-size', '16px')
        ]
    }]).background_gradient(cmap='YlGn',axis=None)
```

```python
moscow21 = opener_of_files('Moscow21.txt')
peter21 = opener_of_files('Petersburg21.txt')
sverdlovsk21 = opener_of_files('Sverdlovsk21.txt')
```

```python
moscow20 = opener_of_files('Moscow20.txt')
peter20 = opener_of_files('Petersburg20.txt')
sverdlovsk20 = opener_of_files('Sverdlovsk20.txt')
```

```python
moscow19 = opener_of_files('Moscow19.txt')
peter19 = opener_of_files('Petersburg19.txt')
sverdlovsk19 = opener_of_files('Sverdlovsk19.txt')
```

```python
moscow18 = opener_of_files('Moscow18.txt')
peter18 = opener_of_files('Petersburg18.txt')
sverdlovsk18 = opener_of_files('Sverdlovsk18.txt')
```

```python
msc21, ptr21, svrd21 = results(moscow21[::2]), results(peter21[::2]), results(sverdlovsk21[::2])
msc20, ptr20, svrd20 = results(moscow20[::2]), results(peter20[::2]), results(sverdlovsk20[::2]) 
msc19, ptr19, svrd19 = results(moscow19[::2]), results(peter19[::2]), results(sverdlovsk19[::2])
msc18, ptr18, svrd18 = results(moscow18[::2]), results(peter18[::2]), results(sverdlovsk18[::2])
```

```python
def crime_frame(name,msc,ptr,svrd):
    crime = pd.DataFrame({'Название преступления': name[1::2],
                        'Москва': msc,
                        'Санкт-Петербург': ptr,
                        'Свердловская область': svrd
                     })
    return crime
```

```python
c1 = crime_frame(moscow21, msc21,ptr21,svrd21)
c2 = crime_frame(moscow20, msc20,ptr20,svrd20)
c3 = crime_frame(moscow19, msc19,ptr19,svrd19)
c4 = crime_frame(moscow18, msc18,ptr18,svrd18)
```

```python
def sum_crimes(frame,year):
    s1,s2,s3 =  frame['Москва'].sum(), frame['Санкт-Петербург'].sum(), frame['Свердловская область'].sum()
    print(f'Общее количество преступлений в {year} составило:\n Москва - {s1}\n Санкт-Петербург - {s2}\n  Свердловская область - {s3}')
```

## Таблица "Показатели преступности России за 2021 год"

```python
table(c1,"Показатели преступности России за 2021 год")
```

```python
c1.plot(figsize=(20,5))
```

```python
sum_crimes(c1,2021)
```

## Заключение

 Мы можем увидеть,что показатели преступлений по Москве,Санкт-Петербургу и Свердловской области после 2018 года имели тенденцию спада,но регистации преступлений увеличивались.
 После 2019 года показатели на первый взгляд стабильно держаться на одном уровне,  В итогах 2021 года можно заметить спад заявлений на регистрацию преступлений,но это не отменяет того факта ,что показатели преступности в каждом пункте упали незначительно ,а в некоторых моментах выросли.
 На данный момент показатели преступности относительно 2018 действительно уменьшились,но все равно на достаточно высоком уровне.
